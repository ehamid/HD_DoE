```python
width = 64
ds_rate = int(img.shape[0] / width)

img = convolve(img, 
    np.array(np.ones((ds_rate,ds_rate))/(ds_rate**2)))[:img.shape[0]:ds_rate,:img.shape[1]:ds_rate]

if max(img.shape) > 64:
    img = img[:64, :64]

f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
magnitude_spectrum = np.log(np.abs(fshift))
real_spectrum = np.log(np.abs(np.real(f)))
imag_spectrum = np.log(np.abs(1e-10 + np.imag(f)))

plt.figure(figsize=(20,20))
plt.subplot(131),plt.imshow(img)
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(real_spectrum)
plt.title('Real Spectrum'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(imag_spectrum)
plt.title('Imaginary Spectrum'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
C = dctn(img, norm='ortho')
# Cshift = np.log(np.abs(np.fft.fftshift(C)))
invC = idctn(C, norm='ortho')
plt.figure(figsize=(20,20))
plt.subplot(131),plt.imshow(img)
plt.title('Input Image')
plt.subplot(132),plt.imshow(np.log(np.abs(C)))
plt.title('DCT')
plt.subplot(133), plt.imshow(invC)
plt.title('iDCT')
plt.show()
```

```python
d = dct(np.eye(img.shape[0]), norm='ortho')
D = np.kron(d,d).T
```

```python
C_vec = D @ np.reshape(img, newshape=[D.shape[0],1])
C_mat = np.reshape(C_vec, img.shape)
plt.subplot(121), plt.imshow(np.log(np.abs(C_mat)))
plt.subplot(122), plt.imshow(np.log(np.abs(C)))
plt.show()
```

```python
import numpy as np
import pywt
from matplotlib import pyplot as plt
from pywt._doc_utils import wavedec2_keys, draw_2d_wp_basis

x = img
shape = x.shape

max_lev = 3       # how many levels of decomposition to draw
label_levels = 3  # how many levels to explicitly label on the plots

fig, axes = plt.subplots(2, 4, figsize=[14, 8])
for level in range(0, max_lev + 1):
    if level == 0:
        # show the original image before decomposition
        axes[0, 0].set_axis_off()
        axes[1, 0].imshow(x, cmap='Blues')
        axes[1, 0].set_title('Image')
        axes[1, 0].set_axis_off()
        continue

    # plot subband boundaries of a standard DWT basis
    draw_2d_wp_basis(shape, wavedec2_keys(level), ax=axes[0, level],
                     label_levels=label_levels)
    axes[0, level].set_title('{} level\ndecomposition'.format(level))

    # compute the 2D DWT
    c = pywt.wavedec2(x, 'db2', mode='periodization', level=level)
    # normalize each coefficient array independently for better visibility
    c[0] /= np.abs(c[0]).max()
    for detail_level in range(level):
        c[detail_level + 1] = [d/np.abs(d).max() for d in c[detail_level + 1]]
    # show the normalized coefficients
    arr, slices = pywt.coeffs_to_array(c)
    axes[1, level].imshow(arr, cmap='Blues')
    axes[1, level].set_title('Coefficients\n({} level)'.format(level))
    axes[1, level].set_axis_off()

plt.tight_layout()
plt.show()
```

```python
beta_vec = pywt.wavedec2(img, 'haar', level=1)
beta_arr, slices = pywt.coeffs_to_array(beta_vec)
d = dct(np.eye(img.shape[0]), norm='ortho')
D = np.kron(d,d).T
```

```python
plt.imshow(beta_arr, cmap = 'Blues')
print(np.sum(np.abs(beta_arr) > 1e-1) / np.product(img.shape))
```

```python
np.sum(np.isclose(beta_arr, 0))
```

```python
X = D.copy()
for j in range(D.shape[0]):
    X_vec = pywt.wavedec2(D[j,:].reshape(img.shape), 'haar', level=2)
    X_arr, _ = pywt.coeffs_to_array(beta_vec)
    X[j,:] = X_arr.reshape([64*64,1])[:,0]
```

```python
y = (X.T @ beta_arr.reshape([64*64,1])).reshape(img.shape)
```

```python
plt.imshow(y)
```

```python
w = np.zeros(img.shape)

for j in range(img.shape[0]):
    w_vec = pywt.wavedec(np.eye(img.shape[0])[:,j], 'haar', level=1)
    w_arr, _ = pywt.coeffs_to_array(w_vec)
    w[:, j] = w_arr
```

```python
plt.imshow(w @ img @ w.T)
print(np.sum(np.abs(w @ img @ w.T) > 1e-10) / np.product(img.shape))
```

```python
np.sum(np.abs(w @ img @ w.T) > 1e-6)
```

```python
W = np.kron(w,w)
```

```python
beta = (W @ img.reshape([W.shape[0],1]))
plt.imshow(beta.reshape(img.shape))
```

```python
np.allclose(W.T @ W, np.eye(W.shape[0]))
```

```python
X = D @ W.T
```

```python
y = X @ beta.reshape([X.shape[0],1])
```

```python
plt.imshow(np.log(np.abs(y.reshape(img.shape))))
plt.show()
```
